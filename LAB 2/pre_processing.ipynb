```python
import pandas as pd
import numpy as np
data = pd.read_csv("Exercise-CarData.csv")
data.head(3)
```

```python
#may perfrom price prediction

data.describe()
```

```python

data = data.iloc[:, 1:]
data
```

### Handle Missing Values

```python
data.isnull().sum()
```

age handle

```python
data["Age"] =  data["Age"].fillna(round(data["Age"].mean())).astype("Int16")
data
```

km handle

```python
#handle ??
data["KM"] = (data["KM"].replace("??",np.nan)).astype(dtype="Int64")
data
```

```python
data["KM"] = data["KM"].fillna(value=data["KM"].mean().__round__())
data["KM"]
```

metcolor handle

```python

data["MetColor"] = data["MetColor"].fillna(value=data["MetColor"].mode()[0]).astype("Int8")
data["MetColor"]
```

handle fueltype

```python
data["FuelType"] = data["FuelType"].fillna(value=data["FuelType"].mode()[0])
data
```

handle hp

```python
data["HP"] = data["HP"].replace("????",np.nan).astype(dtype="Int16")
data["HP"] = data["HP"].fillna(value=data["HP"].mean().__round__())
data
```

handle door

```python
word_to_num = {
  "one" : 1,
  "two" : 2,
  "three" : 3,
  "four" : 4,
  "five" : 5,
  "six": 6,
  "seven": 7,
  "eight" : 8,
  "nine" : 9
}

def transform_word_to_num(x:str|int)->int :
  if isinstance(x,str) and x.lower().strip() in word_to_num:
    return (word_to_num[x])
  return (int)(x)


data["Doors"] = data["Doors"].apply(transform_word_to_num)
```

```python
data
```

### categorical data

```python
data = pd.get_dummies(data=data, columns=["FuelType"],  dtype="int")
data
```

### scaling

```python
from sklearn.preprocessing import StandardScaler

sscaler = StandardScaler()
cols = ["Age", "KM", "HP", "CC", "Doors", "Weight"]

sc = sscaler.fit_transform(data[cols])

df = pd.DataFrame(sc, columns=cols)

data[cols] = df

data
```

```python
data.to_csv("pre-proccesed-data.csv", index=False)
```

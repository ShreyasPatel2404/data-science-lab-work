A. Steps for Data Transformation

```python
# Step 1: Import Libraries

import numpy as np
import pandas as pd
from sklearn.preprocessing import MinMaxScaler,StandardScaler,LabelEncoder,OneHotEncoder
```

```python
# Step 2: Load Data
datasets=pd.read_csv('Exercise-CarData.csv')
print("\nData :\n",datasets)
print("\nData statistics\n",datasets.describe())
```

```python
# Step 3: Seprate Input and Output attributes
x=datasets.iloc[:,:-1].values
y=datasets.iloc[:,-1].values
print("\n\nInput : \n", x)
print("\n\nOutput: \n", y)
```

```python
x_new=datasets.iloc[:,1:3].values
print("\n\nX for transformation : \n", x_new)
```

```python
# Step 4 : Perform scaling on age and salary
scaler=MinMaxScaler()
x_scaled=scaler.fit_transform(x_new)
print("\n\nScaled X : \n", x_scaled)
```

```python
# Step 5 : Perform standardization on age and salary
std=StandardScaler()
x_std=std.fit_transform(x_new)
print("\n\nStandardized X : \n", x_std)
```

B. Steps_for_Handling_Categorical_Data

```python
# Step 4a: Apply LabelEncoder on the data
# to convert country names into numeric values
le = LabelEncoder()
x[ : ,0] = le.fit_transform(x[ : ,0])
print("\n\nInput : \n", x)
```

```python
onehotencoder = OneHotEncoder(handle_unknown='ignore', sparse_output=False)

# Get the 'FuelType' column for one-hot encoding from the original datasets DataFrame
fuel_type_column = datasets['FuelType'].values.reshape(-1, 1)

# Fit and transform the 'FuelType' column
fuel_type_encoded = onehotencoder.fit_transform(fuel_type_column)

# Get feature names for the new one-hot encoded columns
fuel_type_feature_names = onehotencoder.get_feature_names_out(['FuelType'])

# Convert x (current numpy array) to a DataFrame temporarily to easily drop the original 'FuelType' column and concatenate new ones
# We need the column names to accurately drop by name
x_temp_df = pd.DataFrame(x, columns=datasets.columns[:-1])

print(f"Shape of x before One-Hot Encoding FuelType: {x_temp_df.shape}")

# Drop the original 'FuelType' column from the temporary DataFrame
x_temp_df = x_temp_df.drop('FuelType', axis=1)

# Create a DataFrame from the one-hot encoded FuelType array
fuel_type_encoded_df = pd.DataFrame(fuel_type_encoded, columns=fuel_type_feature_names, index=x_temp_df.index)

# Concatenate the modified x_temp_df with the one-hot encoded FuelType DataFrame
x = pd.concat([x_temp_df, fuel_type_encoded_df], axis=1).values

print(f"Shape of x after One-Hot Encoding FuelType: {x.shape}")
print("\nOne-Hot Encoded Input (x) with FuelType:\n", x)
```

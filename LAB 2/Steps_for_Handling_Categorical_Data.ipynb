Steps for Handling Categorical Data

```python
# Step 1: Import Libraries

import numpy as np
import pandas as pd
from sklearn.preprocessing import LabelEncoder,OneHotEncoder
```

```python
# Step 2: Load Data

datasets = pd.read_csv('Data_for_Categorical_Values.csv')
print("\nData :\n",datasets)
print("\nData statistics\n",datasets.describe())
```

```python
# Step 3: Seprate Input and Output attributes
X = datasets.iloc[:, :-1].values
Y = datasets.iloc[:, -1].values
print("\n\nInput : \n", X)
print("\n\nOutput: \n", Y)
```

In Step 3, we separate the dataset into two parts:

Input (X): These are the features used to predict the output.
Output (Y): This is the target variable we want to predict.
This separation is a standard practice in machine learning, allowing us to train a model on the input features to predict the corresponding output.

```python
# Step 4a: Apply LabelEncoder on the data
# to convert country names into numeric values
le = LabelEncoder()
X[ : ,0] = le.fit_transform(X[ : ,0])
print("\n\nInput : \n", X)
```

In Step 4a:

We use a LabelEncoder to convert categorical text data (like country names in our input X) into numerical representations.
Specifically, X[:, 0] refers to the first column of our input data, which contains the 'Country' names. The LabelEncoder assigns a unique integer to each distinct country (e.g., France becomes 0, Germany becomes 1, Spain becomes 2).
This conversion is essential because most machine learning models require numerical input to process data effectively.

In Step 4b, we use a technique called One-Hot Encoding (via pd.get_dummies):

We create new columns for each unique category in the 'Country' column (e.g., 'France', 'Germany', 'Spain').
Each new column will have a True or False value, indicating the presence or absence of that country for each row.
This converts categorical data into a numerical format suitable for machine learning models without implying any order among categories.
Finally, we remove the original 'Country' and 'Purchased' columns and combine the new dummy variables with the rest of our dataset to form the 'Final Data'.

```python
#Use One Hot Encoder from scikit learn
onehotencoder = OneHotEncoder()
#reshape the 1-D country array to 2-D as fit_transform expects 2-D and finally fit the object
x = onehotencoder.fit_transform(datasets.Country.values.reshape(-1,1)).toarray()
x
```

```python
dfOneHot = pd.DataFrame(x, columns = ["Country_"+str(int(i))
for i in range(datasets.shape[1]-1)])
df = pd.concat([datasets, dfOneHot], axis=1) #column
#droping the country column
df= df.drop(['Country'], axis=1)
#printing to verify
print(df.head())
```

dfOneHot = pd.DataFrame(x, columns = ["Country_"+str(int(i)) for i in range(datasets.shape[1]-1)])

x is the output from the OneHotEncoder, which is a NumPy array containing the one-hot encoded representation of the 'Country' column.
pd.DataFrame(x, ...) converts this NumPy array x into a pandas DataFrame.
columns = [...] assigns meaningful column names to this new DataFrame. It dynamically creates names like 'Country_0', 'Country_1', 'Country_2', etc., based on the number of unique countries. datasets.shape[1]-1 is used to determine the number of columns, assuming 'Country' is the only categorical column being one-hot encoded and 'Purchased' is also dropped from the original datasets when considering columns to merge.
df = pd.concat([datasets, dfOneHot], axis=1)

pd.concat() is used to join the datasets DataFrame (which still contains 'Age', 'Salary', 'Purchased', and the original 'Country' column before dropping) and the newly created dfOneHot DataFrame.
axis=1 is crucial here, as it tells pandas to concatenate the DataFrames column-wise, meaning they are joined side-by-side, adding the new one-hot encoded country columns to the datasets DataFrame.
df= df.drop(['Country'], axis=1)

After one-hot encoding the 'Country' column and adding the new dummy variables, the original 'Country' column is no longer needed because its information is now represented by the new one-hot encoded columns.
df.drop(['Country'], axis=1) removes this original 'Country' column from the DataFrame. Again, axis=1 specifies that a column is being dropped.
print(df.head())

This line simply prints the first few rows of the final DataFrame df to allow you to inspect the changes and verify that the one-hot encoding and column dropping have been applied correctly.

```python
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt

n_samples = 300
centers = 4
cluster_std = 0.6

X, y_true = make_blobs(n_samples=n_samples, centers=centers, cluster_std=cluster_std, random_state=42)

plt.figure(figsize=(8, 6))
plt.scatter(X[:, 0], X[:, 1], c=y_true, s=50, cmap='viridis')
plt.title('Initial Distribution of Generated Data')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.grid(True)
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering

# Perform Agglomerative Clustering with 'single' linkage
agg_single = AgglomerativeClustering(n_clusters=4, linkage='single')
agg_single.fit(X)

# Perform Agglomerative Clustering with 'complete' linkage
agg_complete = AgglomerativeClustering(n_clusters=4, linkage='complete')
agg_complete.fit(X)

print("Agglomerative Clustering models (single and complete linkage) fitted successfully.")
```

```python
import matplotlib.pyplot as plt

# Create a figure with subplots
plt.figure(figsize=(18, 6))

# Plotting True Clusters
plt.subplot(1, 3, 1)
plt.scatter(X[:, 0], X[:, 1], c=y_true, s=50, cmap='viridis')
plt.title('True Clusters')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.grid(True)

# Plotting Agglomerative Clustering with Single Linkage
plt.subplot(1, 3, 2)
plt.scatter(X[:, 0], X[:, 1], c=agg_single.labels_, s=50, cmap='viridis')
plt.title('Agglomerative Clustering (Single Linkage)')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.grid(True)

# Plotting Agglomerative Clustering with Complete Linkage
plt.subplot(1, 3, 3)
plt.scatter(X[:, 0], X[:, 1], c=agg_complete.labels_, s=50, cmap='viridis')
plt.title('Agglomerative Clustering (Complete Linkage)')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.grid(True)

plt.tight_layout()
plt.show()

print("Visualizations for true clusters and agglomerative clustering with single and complete linkages generated.")
```

```python
single_linkage_labels = agg_single.labels_
complete_linkage_labels = agg_complete.labels_

print("First 10 cluster labels (Single Linkage):")
print(single_linkage_labels[:10])

print("\nFirst 10 cluster labels (Complete Linkage):")
print(complete_linkage_labels[:10])
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(15, 6))

# Plotting Agglomerative Clustering with Single Linkage
plt.subplot(1, 2, 1)
plt.scatter(X[:, 0], X[:, 1], c=single_linkage_labels, s=50, cmap='viridis')
plt.title('Agglomerative Clustering (Single Linkage)')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.grid(True)

# Plotting Agglomerative Clustering with Complete Linkage
plt.subplot(1, 2, 2)
plt.scatter(X[:, 0], X[:, 1], c=complete_linkage_labels, s=50, cmap='viridis')
plt.title('Agglomerative Clustering (Complete Linkage)')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.grid(True)

plt.tight_layout()
plt.show()

print("Visualizations for agglomerative clustering with single and complete linkages generated.")
```

```python
from scipy.cluster.hierarchy import linkage, dendrogram
import matplotlib.pyplot as plt

# Compute the linkage matrix for 'single' linkage
Z_single = linkage(X, method='single')

# Compute the linkage matrix for 'complete' linkage
Z_complete = linkage(X, method='complete')

# Create a figure with two subplots
plt.figure(figsize=(20, 8))

# Plotting dendrogram for Single Linkage
plt.subplot(1, 2, 1)
dendrogram(Z_single)
plt.title('Dendrogram - Single Linkage')
plt.xlabel('Data Points')
plt.ylabel('Distance')

# Plotting dendrogram for Complete Linkage
plt.subplot(1, 2, 2)
dendrogram(Z_complete)
plt.title('Dendrogram - Complete Linkage')
plt.xlabel('Data Points')
plt.ylabel('Distance')

plt.tight_layout()
plt.show()

print("Dendrograms for single and complete linkages generated successfully.")
```

```python
from sklearn.datasets import load_iris

# Load the Iris dataset
iris = load_iris()

# Access the feature data (X_iris) and target labels (y_iris)
X_iris = iris.data
y_iris = iris.target

print(f"Shape of Iris features (X_iris): {X_iris.shape}")
print(f"Shape of Iris target labels (y_iris): {y_iris.shape}")
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 6))
plt.scatter(X_iris[:, 0], X_iris[:, 1], c=y_iris, s=50, cmap='viridis')
plt.title('Initial Distribution of Iris Dataset')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.grid(True)
plt.show()

print("Scatter plot of Iris dataset generated successfully.")
```

```python
from sklearn.cluster import AgglomerativeClustering

# Perform Agglomerative Clustering with 'single' linkage on Iris dataset
agg_single_iris = AgglomerativeClustering(n_clusters=3, linkage='single')
agg_single_iris.fit(X_iris)

# Perform Agglomerative Clustering with 'complete' linkage on Iris dataset
agg_complete_iris = AgglomerativeClustering(n_clusters=3, linkage='complete')
agg_complete_iris.fit(X_iris)

print("Agglomerative Clustering models for Iris dataset (single and complete linkage) fitted successfully.")
```

```python
import matplotlib.pyplot as plt

# Get the cluster labels from the fitted models
single_linkage_labels_iris = agg_single_iris.labels_
complete_linkage_labels_iris = agg_complete_iris.labels_

plt.figure(figsize=(15, 6))

# Plotting Agglomerative Clustering with Single Linkage for Iris
plt.subplot(1, 2, 1)
plt.scatter(X_iris[:, 0], X_iris[:, 1], c=single_linkage_labels_iris, s=50, cmap='viridis')
plt.title('Agglomerative Clustering (Single Linkage) - Iris Dataset')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.grid(True)

# Plotting Agglomerative Clustering with Complete Linkage for Iris
plt.subplot(1, 2, 2)
plt.scatter(X_iris[:, 0], X_iris[:, 1], c=complete_linkage_labels_iris, s=50, cmap='viridis')
plt.title('Agglomerative Clustering (Complete Linkage) - Iris Dataset')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.grid(True)

plt.tight_layout()
plt.show()

print("Visualizations for Iris dataset clustering with single and complete linkages generated.")
```

```python
from sklearn.metrics import adjusted_rand_score

# Calculate Adjusted Rand Index for Single Linkage Clustering on Iris
ari_single_iris = adjusted_rand_score(y_iris, single_linkage_labels_iris)

# Calculate Adjusted Rand Index for Complete Linkage Clustering on Iris
ari_complete_iris = adjusted_rand_score(y_iris, complete_linkage_labels_iris)

print(f"Adjusted Rand Index (Single Linkage) for Iris: {ari_single_iris:.4f}")
print(f"Adjusted Rand Index (Complete Linkage) for Iris: {ari_complete_iris:.4f}")
```

```python
from scipy.cluster.hierarchy import linkage, dendrogram
import matplotlib.pyplot as plt

# Compute the linkage matrix for 'single' linkage on Iris dataset
Z_single_iris = linkage(X_iris, method='single')

# Compute the linkage matrix for 'complete' linkage on Iris dataset
Z_complete_iris = linkage(X_iris, method='complete')

# Create a figure with two subplots
plt.figure(figsize=(20, 8))

# Plotting dendrogram for Single Linkage (Iris Dataset)
plt.subplot(1, 2, 1)
dendrogram(Z_single_iris)
plt.title('Dendrogram - Single Linkage (Iris Dataset)')
plt.xlabel('Data Points')
plt.ylabel('Distance')

# Plotting dendrogram for Complete Linkage (Iris Dataset)
plt.subplot(1, 2, 2)
dendrogram(Z_complete_iris)
plt.title('Dendrogram - Complete Linkage (Iris Dataset)')
plt.xlabel('Data Points')
plt.ylabel('Distance')

plt.tight_layout()
plt.show()

print("Dendrograms for Iris dataset (single and complete linkages) generated successfully.")
```

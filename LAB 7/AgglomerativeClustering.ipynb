```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
from scipy.cluster import hierarchy
from scipy.spatial import distance_matrix
from matplotlib import pyplot as plt
from sklearn.datasets import make_blobs
```

```python
X1, y1 = make_blobs(
    n_samples=50,
    centers=3,
    cluster_std=1.0,
    random_state=42
)
```

```python
plt.scatter(X1[:, 0], X1[:, 1], c='black', marker='o')
plt.title("Original Data Points")
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering

agglom = AgglomerativeClustering(
    n_clusters=3,
    linkage='single'
)

labels = agglom.fit_predict(X1)
```

```python
x_min, x_max = np.min(X1, axis=0), np.max(X1, axis=0)
X1 = (X1 - x_min) / (x_max - x_min)
```

```python
plt.figure(figsize=(6,4))

for i in range(X1.shape[0]):
    plt.text(
        X1[i, 0], X1[i, 1], 'c',
        color=plt.cm.nipy_spectral(labels[i] / 10.),
        fontdict={'weight': 'bold', 'size': 15}
    )

plt.scatter(X1[:, 0], X1[:, 1], marker='.')
plt.xticks([])
plt.yticks([])
plt.show()
```

```python
linked = hierarchy.linkage(X1, method='single')

plt.figure(figsize=(8, 5))
hierarchy.dendrogram(linked)
plt.title("Dendrogram (Single Linkage)")
plt.xlabel("Data Points")
plt.ylabel("Distance")
plt.show()
```

```python
linked = hierarchy.linkage(X1, method='complete')

plt.figure(figsize=(8, 5))
hierarchy.dendrogram(linked)
plt.title("Dendrogram (Complete Linkage)")
plt.xlabel("Data Points")
plt.ylabel("Distance")
plt.show()
```

```python
linked = hierarchy.linkage(X1, method='average')

plt.figure(figsize=(8, 5))
hierarchy.dendrogram(linked)
plt.title("Dendrogram (average Linkage)")
plt.xlabel("Data Points")
plt.ylabel("Distance")
plt.show()
```

```python
linked = hierarchy.linkage(X1, method='ward')

plt.figure(figsize=(8, 5))
hierarchy.dendrogram(linked)
plt.title("Dendrogram (ward Linkage)")
plt.xlabel("Data Points")
plt.ylabel("Distance")
plt.show()
```
